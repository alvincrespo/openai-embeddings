# OpenAI Embeddings (Getting Started)

Run this notebook using VS Code's Jupyter integration. Make sure your `.venv` interpreter is selected.

```python
from dotenv import load_dotenv
import os
from openai import OpenAI
import json

load_dotenv()

# Set your API key before running (e.g. in the environment as OPENAI_API_KEY)
client = OpenAI()

text = "Hello from OpenAI embeddings!"

response = client.embeddings.create(
    model="text-embedding-3-small",
    input=text
)

embedding = response.model_dump()

# Truncate the embedding for display purposes
emb = embedding["data"][0]["embedding"]
embedding["data"][0]["embedding"] = emb[:3] + ["..."] + emb[-3:]

print(json.dumps(embedding, indent=2))
```
